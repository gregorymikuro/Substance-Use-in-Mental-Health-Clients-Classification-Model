```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd

class DataExaminer:
    def __init__(self, file_path):
        self.data = pd.read_csv(file_path)
        
    def display_first_rows(self, n=5):
        print(self.data.head(n))
        
    def display_summary_statistics(self):
        print(self.data.describe(include='all'))
        
    def display_data_types(self):
        print(self.data.dtypes)
        
    def frequency_tables_and_visuals(self):
        # Define the columns and their corresponding labels
        columns_and_labels = {
            'target-variable': ['Missing/Unknown', 'Yes', 'No'],
            'primary-mental-health-condition': ['Missing/Unknown', 'Trauma/Stressor', 'Anxiety', 'ADHD', 'Conduct', 'Delirium/Dementia', 
                                                'Bipolar', 'Depressive', 'Oppositional Defiant', 'Pervasive Developmental', 
                                                'Personality', 'Schizophrenia/Psychotic', 'Substance Use', 'Other'],
            'age': ['Missing/Unknown', '0-11', '12-14', '15-17', '18-20', '21-24', '25-29', '30-34', '35-39', 
                    '40-44', '45-49', '50-54', '55-59', '60-64', '65+'],
            'ethnicity': ['Missing/Unknown', 'Mexican', 'Puerto Rican', 'Other Hispanic/Latino', 'Not Hispanic/Latino'],
            'race': ['Missing/Unknown', 'American Indian/Alaska Native', 'Asian', 'Black/African American', 
                     'Native Hawaiian/Pacific Islander', 'White', 'Other/Multiracial'],
            'gender': ['Missing/Unknown', 'Male', 'Female'],
            'marital-status': ['Missing/Unknown', 'Never', 'Now', 'Separated', 'Divorced/Widowed'],
            'education': ['Missing/Unknown','Special Education', '0 to 8', '9 to 11', '12 (or GED)', 'More than 12'],
            'employment-status': ['Missing/Unknown', 'Full-time', 'Part-time', 'Employed', 'Unemployed', 'Not in Labor Force'],
            'living-arrangement': ['Missing/Unknown', 'Homelessness', 'Private Residence', 'Other'],
            'veteran-status': ['Missing/Unknown', 'Yes', 'No']
        }

        # Iterate through the columns and labels, and generate the appropriate charts
        for column, labels in columns_and_labels.items():
            if column in self.data.columns:
                if column in ['target-variable', 'gender', 'veteran-status']:
                    self.pie_chart(column, labels)
                else:
                    self.histogram_chart(column, labels)
            else:
                print(f"Column '{column}' is missing, skipping...")
        
    def pie_chart(self, column, labels):
        data = self.data[column].replace([-9, 1, 2], labels)
        data.value_counts().plot.pie(autopct='%1.1f%%', labels=data.value_counts().index)
        plt.title(f'{column} distribution')
        plt.ylabel('')
        plt.show()
    
    def table_chart(self, column, labels):
        data = self.data[column].replace([-9] + list(range(1, len(labels))), labels)
        print(data.value_counts())
        
    def histogram_chart(self, column, labels):
        data = self.data[column].replace([-9] + list(range(1, len(labels))), labels)
        data.value_counts().plot(kind='bar')
        plt.title(f'{column} distribution')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.show()


        
"""    def plot_state_fips_map(self):
        # Assuming we have a shapefile for US states
        usa = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
        usa = usa[(usa.continent == 'North America') & (usa.name != 'Canada') & (usa.name != 'Greenland')]

        state_fips = self.data['state-fips-code'].value_counts().reset_index()
        state_fips.columns = ['state-fips-code', 'count']

        usa = usa.merge(state_fips, how='left', left_on='iso_a2', right_on='state-fips-code')
        
        usa.plot(column='count', cmap='OrRd', legend=True)
        plt.title('State FIPS Code Distribution')
        plt.show()
        
"""

# Create an instance of DataExaminer
data_examiner = DataExaminer('data/extracted-data.csv')
data_examiner.display_first_rows()
data_examiner.display_summary_statistics()
data_examiner.display_data_types()
data_examiner.frequency_tables_and_visuals()
```
